```python
def from_json(nom_fichier):
    liste_csv = []
    try:
        file = open(nom_fichier, "r", encoding="utf8")
        for line in file:
            line = line.split("\n")[0]
            liste_csv.append(eval(line))
        file.close()
    except Exception as err:
        print(err)
    return liste_csv
```

```python
from_json("extrait.txt")
```

```python
json = from_json("extrait.txt")

def formattage(data):
    for film in data:
        acteurs = film['cast']
        for i in range(len(acteurs)):
            acteur = acteurs[i].strip('[]')
            acteurs[i] = acteur
formattage(json)
```

```python
import networkx as nx

G = nx.Graph()

def genere_graph(data):
    for film in data:
        acteurs = film['cast']
        for i in range(len(acteurs)):
            for j in range(i+1, len(acteurs)):
                G.add_edge(acteurs[i], acteurs[j])

genere_graph(json)
```

```python
nx.draw(G, with_labels=True)
```
